```python
# generate data from a line plus noise and save it in a pkl file
import numpy as np
import matplotlib.pyplot as plt
import pickle as pkl

# generate data
np.random.seed(0)
n = 100
x = np.random.uniform(low=-1.0, high=1.0, size=(n,2))
y = (x[:, 1] > 2 * x[:, 0]).astype(int)
# # add some label noise to y
# rand_mask = np.random.uniform(size=n) < 0.1
# y[rand_mask] = abs(1 - y[rand_mask])
# visualize the data
plt.figure()
plt.scatter(x[y==0, 0], x[y==0, 1], c='r', marker='x', label='y=0')
plt.scatter(x[y==1, 0], x[y==1, 1], c='b', marker='o', label='y=1')

with open("linear_data.pkl", "wb") as f:
    pkl.dump((x, y), f) 
```

```python
# generate some circular data
n = 1000
x = np.random.uniform(low=-1.0, high=1.0, size=(n,2))
y = (np.linalg.norm(x, axis=1) < 0.8).astype(int)

# visualize the data
plt.figure(figsize=(5,5))
plt.scatter(x[y==0, 0], x[y==0, 1], c='r', marker='x', label='y=0')
plt.scatter(x[y==1, 0], x[y==1, 1], c='b', marker='o', label='y=1')
plt.show()

with open("secret_data.pkl", "wb") as f:
    pkl.dump((x, y), f) 
```
