```python
import matplotlib.pyplot as plt
# Lets do descent on a quadratic function
def f(x):
    return x**2

def grad_f(x):
    return 2 * x

x_init = 10
lr = 0.01
beta = 0.8
f_history = []
x = x_init
momentum_value = lr * grad_f(x)
for i in range(100):
    f_history.append(f(x))
    momentum_value = beta * momentum_value + (1 - beta) * grad_f(x)
    x -= lr * momentum_value

    
x_init = 10
momentum_value = 0
lr = 0.01
beta = 0.98
f_history_sgd = []
x = x_init
for i in range(100):
    f_history_sgd.append(f(x))
    momentum_value = grad_f(x)
    x -= lr * momentum_value
    
plt.figure()
plt.plot(f_history, label='momentum')
plt.plot(f_history_sgd, label='sgd')
plt.legend()
```
